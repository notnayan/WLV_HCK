##**1.2** Data Transformation:

Write code to transform variables according to the following instructions:

1. ”School”, ”sex”, ”address”, ”schoolsup”, ”famsup”, ”activites”, ”nursery”, ”internet”, and ”romantic”
into binary: 0 or 1 (create new columns without overwriting the existing ones).


```python
# Columns list
columnBinary = ["school", "sex", "address", "schoolsup", "famsup", "activities", "nursery", "internet", "romantic"]

# Making a new column by mapping each of the following: "Yes to 1 and No to 0," "GP to 0 and MS to 1," "M to 0 and F to 1," "U to 0 and R to 1," through the given column.
for column in columnBinary:
    df[column+"_newbinary"] = df[column].map({"yes":1, "no":0, "GP":0, "MS":1, "M":0, "F":1, "U":0,"R":1})
df
```

2. ”Medu”, ”Fedu”, ”reason”, ”guardian”, ”studytime”, ”freetime”, and ”health” into ordinal numbers based
on the number cases in the data set (create news columns without overwriting the existing ones).

```python
# Column List
columnsOrdinal = ['Medu', 'Fedu', 'reason', 'guardian', 'studytime', 'freetime', 'health']

# Encoding into new ordinal numbers while looping through the provided columns
for column in columnsOrdinal:
    df[column + '_newordinal'] = pd.Categorical(df[column]).codes
df
```

3. Convert column ”age” to interval datatype. i.e. Create a new column name category age whose values
should be based on the frequency in the column ”age”, You can create categorical data with following
interval.

  (a) interval1: [15-17]

  (b) interval2: [18-20]

  (c) interval3: [21-all]

```python
# Define the given intervals for the categories_age column
givenIntervals = [15, 17, 20, np.inf]

# Create the age categories using pd.cut and include lowest to take lower bound into account
df['categoryAge'] = pd.cut(df['age'], givenIntervals, labels=['[15-17]', '[17-20]', '[20-all]'], include_lowest=True)

df
```

4. Create a new column name passed (yes or no) whose values should be based on the values present in the
G3 column (>= 8–yes, < −no).

```python
# Using pandas pandas.Series.apply() and passing an anonymous function that returns "yes" or "no" depending on whether G3 is more than or equal to 8 or not.
df['passed'] = df['G3'].apply(lambda x: "yes" if x >=8 else "no")
df
```

1. Write a code to show histogram plots and boxplots to visualize the distribution of the variables
”age”, ”absence”, and ”G3”. Interpret the results and comment on the distribution of each variable.
Write your answer in text cell of jupyter notebook.

```python
# Column List
columns = ['age', 'absences', 'G3']

# Show histograms for the selected columns
for column in columns:
  
  # Labeling the histogram
  plt.title("Distribution of " + column)
  plt.xlabel(column.title())
  plt.ylabel("Frequency")

  # Showing the histogram
  plt.hist(df[column])
  plt.show()
```

```python
# Show boxplots for the selected columns
for column in columns:

  # Labeling the boxplot
  plt.title("Distribution of " + column)
  plt.xlabel(column.title())
  plt.ylabel("Frequency")
  
  # Showing the boxplot
  sns.boxplot(data=df[column])
  plt.show()
```

I. The frequency of students is indicated by the y-axis while the x-axis displays age. The graph displays the number of students in each age group.

II. The frequency of students is shown on the y-axis while the x-axis displays absences. The bar graph displays the percentage of missing students for each day.

III. The final grade is shown on the x-axis, while the frequency of students is shown on the y-axis. The graph displays the final scores as well as the percentage of students that achieved them.

2. Write a code to show a bar graph of the total number of students who passed the final term grouped
according to the school that they belong to. Use proper labels in the graph and interpret the results.

```python
# Count the number of students who passed and group them by their schools using the previously created "Passed" column
schoolCount = df.groupby(['school'])['passed'].apply(lambda x: x[x == 'yes'].count())

# Plot a bar graph with the count
schoolCount.plot(kind='bar')

# Labeling the bar chart
plt.title('Number of Students Who Passed Final Term VS School')
plt.xlabel('Schools')
plt.ylabel('Number of Students Who Passed Final Term')

# Show the bar graph plot
plt.show()
```

-> The number of students who passed their last term is depicted in the graph above each school. Here, the school is represented by the x-axis, and the number of students who passed the last term is represented by the y-axis. The frequency of students from each school that passed their last term is shown on the graph. It is obvious that there are more GP graduates than MS grads.

3. Write a code to show a pie chart of analysis of students’ grades as per their school. Use proper labels
in the graph and interpret the results.

```python
# Group the data by school
groupedDF = df.groupby('school')

# Create pie chart for each school
for name, group in groupedDF:
    group['G3'].value_counts().plot.pie(title='Grade of Students of ' + name, figsize=(11,11))
    plt.legend(title='Grades', loc='upper right', fontsize=8)
    plt.show()
```

I. The first pie chart displays the distribution of school GP final grades from 0 to 20. From the char it is obvious that the number of pupils who received 10 is at its highest, but the grade 4 is at its lowest.

II. The pie-chart displays the distribution of school MS final grades from 0 to 20. It is obvious that the number of pupils who received 10 is the maximum, although the lowest and equal grades are 6, 14, and 19.

At the end we can clearly see that the marks are more evenly distributed in school MS than in GP.

4. Write a code to show a bar graph with the relation of the father’s occupation with the grade of the
student. Use proper labels in the graph and interpret the results.

```python
# Group the data by father's occupation and calculate the mean grade for each occupation
meanGrade = df.groupby("Fjob")["G3"].mean()

# Plotting the bar graph
meanGrade.plot.bar(title="Father's Occupation VS Grade")

# Labelling the graph
plt.xlabel("Father's Occupation")
plt.ylabel("Mean Grade")

plt.show()
```

-> A graph showing the relationship between Father's Occupation and Grade is shown above. Here, the y-axis shows the mean student grade while the x-axis shows the father's occupation, such as staying at home or working in the health sector etc. The purpose of the graph plot is to demonstrate how the father's employment impacts the students' grades. The average student grade is calculated, and it is compared with the profession of the father.
